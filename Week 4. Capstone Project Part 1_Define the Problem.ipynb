## Introduction

With the development of the global economy, more and more people change their places of residence due to the needs of study and work. In this project, we will take a customer as an example to discuss how to use data analysis to find a satisfactory residence for the customer.

In this case, the client moved from Toronto to New York City. When he is satisfied with his current location on the west side of Toronto, he is looking for similar neighborhoods in New York City, such as shops and restaurants. We will use data science techniques to help clients identify the neighborhoods in New York City that are most similar to his current residential area. 

This project is not only suitable for our customers, but also for customers who are interested in finding similar neighbors in other cities around the world.

## Data Description

The analysis will be conducted into 3 parts:

#### Part 1: Explore the Client 's Current Neighborhood in Toronto
Web scraping techniques will be applied to retrieved Toronto neighborhood data from Wikipedia"https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M". After building a dataframe, we will add the geographical coordinates of a given postal code. Then, we will explore and analyze the given Neighborhoods in Toronto:
1. using Folium to visualize the west side of Toronto
2. utilizing the Foursquare API to explore the neighborhoods and segment them.

#### Part 2: Explore the Neighborhoods in New York City
The NYC Dataset contains 5 boroughs and the neighborhoods that exist in each borough as well as the latitude and longitude coordinates of each neighborhood, which can be downloaded from https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DS0701EN-SkillsNetwork/labs/newyork_data.json.
Same analyses from part one will be applied to the NYC neighborhood dataset. In addition to that, we will use k-means to cluster the neighborhood of NYC.

#### Part 3: Compare the Neighborhood Clusters of Toronto and NYC 
Finally, we will compare the clusters of NYC with the neighborhood of Toronto and find out the most similar neighborhoods in NYC to the client's current residential neighborhoods in Toronto.
